```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('magic04.data')
print(df.head())
```

```python
df.columns
```

```python
#Taking column names from magic04.names
column_names = [
    'fLength', 'fWidth', 'fSize', 'fConc', 'fConc1',
    'fAsym', 'fM3Long', 'fM3Trans', 'fAlpha', 'fDist', 'class'
]
df.columns = column_names
df['class'] = df['class'].astype(str).str.strip()
print(df.head())
```

```python
missing_counts = df.isnull().sum()
print("Missing values per column:")
print(missing_counts)
```

```python
duplicate_count = df.duplicated().sum()
print("Number of duplicate rows:", duplicate_count)
```

```python
df = df.drop_duplicates()
duplicate_count = df.duplicated().sum()
print("Number of duplicate rows:", duplicate_count)
```

```python
print("Checking box plot of fLength to indentify outliers")
plt.boxplot(df['fLength'])
plt.show()
print("Checking box plot of fAsym to indentify outliers")
plt.boxplot(df['fAsym'])
plt.show()
```

```python
for col in ['fLength', 'fWidth', 'fSize', 'fConc', 'fConc1', 'fAsym', 'fM3Long', 'fM3Trans', 'fAlpha', 'fDist']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower, upper = Q1 - 1.5*IQR, Q3 + 1.5*IQR
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    print(f"{col}: detected", len(outliers), "outliers; bounds:", lower, upper)
    df = df[(df[col] >= lower) & (df[col] <= upper)]
```

```python
print("Checking box plot of fLength to see result")
plt.boxplot(df['fLength'])
plt.show()
print("Checking box plot of fAsym to see result")
plt.boxplot(df['fAsym'])
plt.show()
```

```python
df['class'] = df['class'].map({'g': 1, 'h': 0})
df.sample(3)
```

```python
from sklearn.model_selection import train_test_split
model_df = df.copy()
print(f"DataFrame shape before splitting: {model_df.shape}")

X = model_df.drop(columns=['class'])
y = model_df['class']

print("\nFeatures (X) preview:")
print(X.head())
print("\nTarget (y) preview:")
print(y.head())

# Perform the train-test split
# test_size=0.20 means 20% of the data will be used for testing, 80% for training.
# random_state ensures reproducibility, so you get the same split every time.
# stratify=y ensures that the proportion of 'Survived' (0s and 1s) is similar in both train and test sets.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42, stratify=y)

print("\n--- Data Splitting Complete ---")
print(f"Shape of X_train: {X_train.shape}")
print(f"Shape of X_test: {X_test.shape}")
print(f"Shape of y_train: {y_train.shape}")
print(f"Shape of y_test: {y_test.shape}")

print("\nProportion of Survived in original dataset:")
print(y.value_counts(normalize=True))
print("\nProportion of Survived in training set:")
print(y_train.value_counts(normalize=True))
print("\nProportion of Survived in test set:")
print(y_test.value_counts(normalize=True))
```

```python
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

numerical_features = X_train.select_dtypes(include=np.number).columns.tolist()

print(f"Numerical features to scale: {numerical_features}")

scaler = StandardScaler()

scaler.fit(X_train[numerical_features])

X_train_scaled_numerical = scaler.transform(X_train[numerical_features])
X_test_scaled_numerical = scaler.transform(X_test[numerical_features])

X_train_scaled = pd.DataFrame(X_train_scaled_numerical, columns=numerical_features, index=X_train.index)
X_test_scaled = pd.DataFrame(X_test_scaled_numerical, columns=numerical_features, index=X_test.index)

print("\n--- X_train after scaling numerical features (head()) ---")
print(X_train_scaled.head())

print("\n--- Descriptive statistics of scaled X_train numerical features ---")
print(X_train_scaled.describe())

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(X_train['fLength'], kde=True)
plt.title('fLength Distribution Before Scaling')
plt.subplot(1, 2, 2)
sns.histplot(X_train_scaled['fLength'], kde=True)
plt.title('fLength Distribution After Standardization')
plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

X_train_processed = X_train_scaled
X_test_processed = X_test_scaled

print("--- Training Logistic Regression Model ---")

model = LogisticRegression(random_state=42, solver='liblinear')

model.fit(X_train_processed, y_train)

print("Model training complete!")

print("\n--- Making Predictions on the Test Set ---")

y_pred = model.predict(X_test_processed)

y_pred_proba = model.predict_proba(X_test_processed)[:, 1]

print("Predictions made successfully!")
print("Sample of predicted classes (first 10):", y_pred[:10])
print("Sample of predicted probabilities (first 10):", y_pred_proba[:10])

print("\n--- Evaluating Model Performance ---")

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")

conf_matrix = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(conf_matrix)

plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['h(0)', 'g(1)'],
            yticklabels=['h(0)', 'g(1)'])
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix for Logistic Regression')
plt.show()

clf_report = classification_report(y_test, y_pred)
print("\nClassification Report:\n", clf_report)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

print("--- Training Decision Tree Classifier Model ---")

dt_model = DecisionTreeClassifier(max_depth=10, random_state=42)

dt_model.fit(X_train_processed, y_train)

print("Model training complete!")

print("\n--- Making Predictions on the Test Set ---")

y_pred_dt = dt_model.predict(X_test_processed)

y_pred_proba_dt = dt_model.predict_proba(X_test_processed)[:, 1]

print("Predictions made successfully!")
print("Sample of predicted classes (first 10):", y_pred_dt[:10])
print("Sample of predicted probabilities (first 10):", y_pred_proba_dt[:10])

print("\n--- Evaluating Model Performance ---")

dt_accuracy = accuracy_score(y_test, y_pred_dt)
print(f"Accuracy: {dt_accuracy:.4f}")

dt_conf_matrix = confusion_matrix(y_test, y_pred_dt)
print("\nConfusion Matrix:")
print(dt_conf_matrix)

plt.figure(figsize=(6, 4))
sns.heatmap(dt_conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['h(0)', 'g(1)'],
            yticklabels=['h(0)', 'g(1)'])
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix for Decision Tree')
plt.show()

dt_clf_report = classification_report(y_test, y_pred_dt)
print("\nClassification Report:\n", dt_clf_report)
```
